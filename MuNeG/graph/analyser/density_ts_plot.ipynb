```python
#real data
frame = pd.DataFrame()
frame = pd.read_csv('D:\\pycharm_workspace\\multiplex\\MuNeG\\results\\real_sw\\lustre_results\\combined_real_sw_density_with_homogenity.csv',index_col=None, header=0)
nrOfLayers = 6
```

```python
frame['fold'] = frame['fold'].map({0:0, 2: 50, 3: 66, 4:75, 5:80, 10:90, 20:95})
frame = frame.drop(frame[frame.fold == 66].index)
frame = frame.drop(frame[frame.fold == 75].index)
frame = frame.drop(frame[frame.fold == 80].index)
frame = frame.drop(frame[frame.fold == 90].index)
frame = frame.drop(frame[frame.fold == 95].index)
```

```python
frame = pd.melt(frame, id_vars=['density', 'rep', 'homogenity'],value_vars=["LR","SF","WL","FG","RF","SCF","FCF"])
frame["variable"] = pd.Categorical(frame['variable'], ["LR","SF","WL","FG","RF","SCF","FCF"])
```

```python
frame["density"] = frame["density"] / float(nrOfLayers)
```

```python
for group in xrange(0, 1001, 1):
    group_name = float(group) / 1000.0
    mask = frame.density <= group_name
    column_name = 'density'
    frame.loc[mask, column_name] = str(group_name)

frame = frame.sort_values(by=["variable","density"])
```

```python
frame = frame.groupby(['density', 'variable'])["value"].mean().reset_index().dropna()
```

```python
sns.set(font='monospace', font_scale=2, palette="deep", style='whitegrid')
```

```python
frame.rename(columns = {'variable':'method'}, inplace = True)
frame['rep']=1
```

```python
frame = frame.assign(degree = lambda x: x.density)
frame.degree = frame.degree.astype(float)
frame = frame.assign(degree = lambda x: x.degree * 90)
```

```python
fig, ax = plt.subplots()
```

```python
ax = sns.tsplot(data=frame,
               time='degree',
               value='value',
               unit='rep',
               condition='method',
               err_style="boot_traces", n_boot=500)
ax.set_ylim(ymin=0.0)
ax.set_ylim(ymax=1.0)
ax.set_xlim(xmax=5)
plt.setp(ax.lines,linewidth=3) 
plt.setp(ax.legend_.get_lines(),linewidth=3)
plt.title('AuC classification results for 50% of known nodes')
plt.xlabel('Mean degree')
plt.ylabel('Area under ROC-Curve for different density')
```

```python
ax2 = ax.twinx()
homogeneity_line = plt.axvline(x=1.77)
homogeneity_line.set_color("black")
homogeneity_line.set_linestyle("dashed")
homogeneity_line.set_lw(1)
ax2.grid(False)
ax2.yaxis.set_major_formatter(plt.NullFormatter())
```
